```python
'''student = [[1,'Steve'],[2,'David'],[3,'Aryan']] ->  ['student_id', 'name'])
marks = [[1,'pyspark',90],[1,'sql',100],[2,'sql',70],[2,'pyspark',60],[3,'sql',30],[3,'pyspark',20]] -> ['student_id','subject_name','marks'])
 
Calculate the % Marks for each student. Each subject is of 100 marks. Create a result column by following the below condition
 
% Marks greater than or euqal to 70 then 'Distinction'
% Marks range between 60-69 then 'First Class'
% Marks range between 50-59 then 'Second Class'
% Marks range between 40-49 then 'Third Class'
% Marks Less than or equal to 39 then 'Fail'
'''

from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark = SparkSession.builder.appName("Create DataFrame").getOrCreate()
student = [[1,'Steve'],[2,'David'],[3,'Aryan']]
marks = [[1,'pyspark',90],[1,'sql',100],[2,'sql',70],[2,'pyspark',60],[3,'sql',30],[3,'pyspark',20]]
student_df = spark.createDataFrame(student,['student_id','name'])
marks_df = spark.createDataFrame(marks,['student_id','subject_name','marks'])
df = marks_df.groupBy('student_id').agg(sum('marks').alias('total_marks'))
df1 = student_df.join(df,student_df.student_id == df.student_id,'inner')
df2 = df1.withColumn('result',when(col('total_marks') >= 70,'Distinction').when((col('total_marks') >= 60) & (col('total_marks') < 70),'First Class').when((col('total_marks') >= 50) & (col('total_marks') < 60),'Second Class').when((col('total_marks') >= 40) & (col('total_marks') < 50),'Third Class').when(col('total_marks') <= 39,'Fail'))
df2.show()
display(df)
```
