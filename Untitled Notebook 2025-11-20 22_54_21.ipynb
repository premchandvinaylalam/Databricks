```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Create DataFrames").getOrCreate()
emp_data = [(1, "Smith", -1, 2000), (2, "Rose", 1, 3000), (3, "Williams", 1, 4000), (4, "Jones", 2, 4000), (5, "Brown", 2, 5000)]
columns = ["emp_id", "emp_name", "raking", "bones"]
df = spark.createDataFrame(emp_data,columns)
df.show()
```
