```python
spark = SparkSession.builder.appName("Create DataFrame").getOrCreate()
empdata = [(1,"sravan","vijayawada",100000),(2,"ramana","vijayawada",200000),(3,"deepika","hyderabad",300000),(3,"deepika","hyderabad",300000)]
columns = ["emp_id","emp_name","address"    ,"salary"]
df = spark.createDataFrame(empdata,columns)
display(df)
df_distinct = df.dropDuplicates(["emp_name"])
df_distinct.show()
```

```python
df_distinct1 = df.distinct()
df_distinct1.show()
```

```python
from pyspark.sql import *
from pyspark.sql.functions import row_number,rank,dense_rank
from pyspark.sql.window import Window
df_window = Window.partitionBy("address").orderBy("address")
df_rownumber = df.withColumn("row_number",row_number().over(df_window))
df_rank = df.withColumn("rank",rank().over(df_window))
df_dense_rank = df.withColumn("dense_rank",dense_rank().over(df_window))
display(df_dense_rank)
display(df_rank)
display(df_rownumber)
```
