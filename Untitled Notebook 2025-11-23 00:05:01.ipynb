```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName(
    "createDataFrame"
).getOrCreate()

cust = [
    (1, "john", "2024-01-01:21:00:00"),
    (2, "mary", "2024-01-02:22:00:00"),
    (3, "peter", "2024-01-03:23:00:00"),
    (4, "jane", "2024-01-04:24:00:00")
]

columns = ["id", "name", "date"]

df = spark.createDataFrame(
    data=cust,
    schema=columns
)

display(df)
```

```python
df1 = spark.read.format("csv").option("header",True).option("path","/Volumes/workspace/default/customers/customer_daily_duplicate.csv").load()
display(df1)
```

```python
from pyspark.sql import *
from pyspark.sql.window import *
from pyspark.sql.functions import *
win = Window.partitionBy("customer_id").orderBy(desc("updated_timestamp"))
df2 = df1.withColumn("rn",row_number().over(win)).withColumn("rank",rank().over(win)).withColumn("denk",dense_rank().over(win)).filter("rn == 1").drop("rn")
display(df2)
```
